Hi Guys Cristopher Nolan is here. Don't talk about this, but I am making a new Batman movie and I need your help to design the new Batmobile. For this ocassion I would like you to design it as several Python Classes.

- Design the main Batmobile class with the following attributes: Pilot (It could be Batman or Robin), number of batrockets loaded, number of pem (this are electromagnetic pulses to counter the Joker weapons) uses left, liters of gas left in the batmobile

- Add the following methods to the Batmobile class: A method to accelerate the batmobile and drive for some distance, a method to activate the turbo and drive for some distance faster than the normal method, a method to launch a rocket to a Joker colleague, a method to launch a pem use.

- Create the following classes to interact with the Batmobile: Joker, Joker Villain, Rocket, PEM Pulse.

- The Joker and Jokker Villain class should have a parameter called KO which will turn to True when they receive a Rocket. They also have a parameter called Weapon which should turn to false if they receive a PEM Pulse. 

- The Rocket and PEM Pulse are the ammunition of the Batmobile, the car can only use the weapons if it has ammunition in the arsenal.

- Finally draw a diagram of the classes using diagram.io showing how they interact.

```

class Rocket:
    def __init__(self, units=1):
        self.units = units

class PEM_Pulse:
    def __init__(self, units=1):
        self.units = units

class Batmobile:
    def __init__(self, pilot, list_rockets, list_pems, gas_liters):
      # Variables init
        self.pilot = pilot #Name of pilot
        self.num_pems = 0  #Totalizer num pems
        self.num_rockets = 0 #Totalizer num rockets
        self.list_rockets = list_rockets #List of configuration rockets
        self.list_pems = list_pems #List of configuration pems
        self.gas_liters = gas_liters #Gas liters available
      #Sequence init  
        print(f"Inicializing Batmobile ...\nChecking Status:\n")

        state = self._check() 
        if state : 
          print("Batmobile Enabled. Ready for charge ...\n") 
        else:
          raise Exception("Batmobile Disabled. Try Again ...\n")  
        print("Charging ammunition ...")
        self._charge_ammunition()
        print(f"Gas liters: {self.gas_liters}\n")
        
    def _check(self):
      # Hidden Method for checking data and ammunition type 
      _enable= False
      if (self.pilot == "Batman" or self.pilot== "Robin"):
        print(f"Welcome Mr.{self.pilot}\n")
        _enable = True
      else:
        print(f"Mr.{self.pilot}: You are a unauthorized pilot. Get out of my Batmobile!!\n")
        raise Exception("Batmobile Disabled. Try Again ...\n")
        _enable = False
        
      for missile in self.list_rockets:
        if isinstance(missile, Rocket):
          _enable = True
          print("Rockets enabled --> OK\n")
        else:
          _enable = False
          raise Exception("This ammunition is not allowed\n")
            
      for pulse in self.list_pems:
        if isinstance(pulse, PEM_Pulse):
          _enable = True
          print("Pems enabled --> OK\n")
        else:
          _enable = False
          raise Exception("This ammunition is not allowed\n")
      
      return(_enable)

    def _charge_ammunition(self):
        # Hidden Method for charging Batmobile with ammo (a kind of initializer): 
    
        num_missile = 0
        for missile in self.list_rockets:
            num_missile += missile.units
        self.num_rockets = num_missile
        num_pulses = 0
        for pulse in self.list_pems:
            num_pulses += pulse.units
        self.num_pems = num_pulses
        print(f"Ammunition charged:\nBatrockets: {self.num_rockets}\nPEM: {self.num_pems}\n ")
    
    def status_ammo(self):
      #Method to print ammunition status
      print(f"Ammunition Status:\nBatrockets: {self.num_rockets}\nPEM: {self.num_pems}\n ")



    def accelerate(self, t_acc=1):
        #Method to advance at normal Speed(avg)= 2Km / min. Manage the advance of the Batmobile, gas liters consumption and distance traveled.
        #t_acc : time in minutes that Batmobile will be in a accelerate speed
        #consumption: comsumption liters of gas per minutes constant (l/min)
        consumption= 1
        liters_left = self.gas_liters
        if liters_left > (t_acc*consumption):
            liters_left += -(t_acc*consumption)
            self.gas_liters = liters_left
            print(f"Distance traveled {t_acc*2}Km\n{self.gas_liters}L of gas left ")
        else:
            print(f"Batmobile need more gas: Gas liters left= {liters_left}L")     
        
        return 

    def turbo(self, t_turbo=1):        
        #Method to advance at turbo Speed(avg)= 4Km / min. Manage the advance of the Batmobile, gas liters consumption and distance traveled.
        #t_turbo : time in minutes that Batmobile will be in turbo speed
        #consumption: comsumption liters of gas per minutes constant (l/min)
        consumption=2
        liters_left = self.gas_liters
        if liters_left > (t_turbo*consumption):
            liters_left += -(t_turbo*consumption)
            self.gas_liters = liters_left
            print(f"Distance traveled {t_turbo*4}Km\n{self.gas_liters}L of gas left ")
        else:
            print(f"Batmobile need more gas: Gas liters left= {liters_left}L")        
        
        return 

    def launch_rocket(self, enemy, rockets_launched=1): 
        #Select Enemy and rockets launched.
    
        rockets_left = self.num_rockets
        if isinstance(enemy, JokerVillain):
            pass
        elif isinstance(enemy, Joker):
            pass
        else:
            raise Exception("Civil Casualties? Are U sure?")
        if enemy.KO == True:
            print(f"Enemy {enemy.__class__.__name__} is KO.Not waste your ammunition.")
        else:
            if rockets_left >= rockets_launched:
                rockets_left -= rockets_launched
                enemy.KO = True
                self.num_rockets = rockets_left
                print(f"Launched {rockets_launched} rockets. {self.num_rockets} rockets left. Enemy {enemy.__class__.__name__} is KO")
            else:
                print("Rocket-Type Amo Empty: You Cannot Launch Rockets!!!")
        
        return

    def launch_pem(self, enemy, pems_launched=1): 
        #Select Enemy and Pems launched
        
        pems_left = self.num_pems
        if isinstance(enemy, JokerVillain):
            pass 
        elif isinstance(enemy, Joker):
            pass
        else:
            raise Exception("Civil Casualties? Are U sure?")
        if enemy.KO==True: 
            print(f"Enemy {enemy.__class__.__name__} is KO.Not waste your ammunition.")
        else:
            if pems_left >= pems_launched:
                pems_left -= pems_launched
                enemy.Weapon = False
                self.num_pems = pems_left  
                print(f"Launched {pems_launched} Pems.--> {self.num_pems} PEMs left. Enemy {enemy.__class__.__name__} cannot shoot")
            else:
                print("PEM-Type Amo Empty: You Cannot Launch PEMs!!!")
        
              
        return

class Joker:
    def __init__(self, KO=False, Weapon=True):
        self.KO = KO
        self.Weapon = Weapon        
        

class JokerVillain:
    def __init__(self, KO=False, Weapon=True):
        self.KO = KO
        self.Weapon = Weapon
      
```
